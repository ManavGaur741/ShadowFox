##  Library Overview

###  Matplotlib
Matplotlib is a foundational plotting library for Python. It provides full control over figure elements and supports a wide range of plot types.

- **Use Cases**: Scientific visualization, static charts for publication.
- **Strengths**: Highly customizable, mature ecosystem.
- **Weaknesses**: Verbose syntax, steeper learning curve.

###  Seaborn
Seaborn is built on top of Matplotlib and offers a cleaner syntax for attractive and complex visualizations with fewer lines of code.

- **Use Cases**: Statistical plots, exploratory data analysis.
- **Strengths**: Beautiful default styles, easy syntax.
- **Weaknesses**: Limited low-level customization compared to Matplotlib.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Load dataset
iris_data = pd.read_csv(r"D:\ShadowFox\synthetic_iris_dataset.csv")
iris_data.head()
```

###  Line Plot – Matplotlib
**Use Case**: Shows sepal length trend for first 10 entries.

```python
plt.plot(iris_data['sepal_length'][:10], marker='o')
plt.title('Line Plot - Sepal Length (First 10 Samples)')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Length')
plt.grid(True)
plt.show()
```

###  Bar Chart – Seaborn
**Use Case**: Shows count of each species.

```python
species_count = iris_data['species'].value_counts().reset_index()
species_count.columns = ['species', 'count']
sns.barplot(x='species', y='count', data=species_count)
plt.title('Bar Chart - Species Count')
plt.show()
```

###  Histogram – Matplotlib
**Use Case**: Distribution of petal width.

```python
plt.hist(iris_data['petal_width'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram - Petal Width')
plt.xlabel('Petal Width')
plt.ylabel('Frequency')
plt.show()
```

###  Pie Chart – Matplotlib
**Use Case**: Proportion of species in dataset.

```python
species_counts = iris_data['species'].value_counts()
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%')
plt.title('Pie Chart - Species Distribution')
plt.show()
```

###  Scatter Plot – Seaborn
**Use Case**: Shows relation between sepal and petal length.

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris_data)
plt.title('Scatter Plot - Sepal vs Petal Length')
plt.show()
```

###  Box Plot – Seaborn
**Use Case**: Compare sepal length distribution by species.

```python
sns.boxplot(x='species', y='sepal_length', data=iris_data)
plt.title('Box Plot - Sepal Length by Species')
plt.show()
```

##  Comparison of Libraries

| Feature              | Matplotlib                     | Seaborn                           |
|----------------------|--------------------------------|-----------------------------------|
| Syntax               | Low-level, verbose             | High-level, concise               |
| Customization        | Very high                      | Moderate                          |
| Default Aesthetics   | Basic                          | Attractive and publication-ready  |
| Interactivity        | Limited (needs extra tools)    | Built-in via Matplotlib           |
| Best Use Cases       | Full control, fine-tuning      | Quick EDA, statistical plots      |
| Performance          | Fast with small/medium data    | Optimized for tidy DataFrames     |

Choose **Matplotlib** when you need full control. Use **Seaborn** for quick and beautiful plots.
